```python
import re
import math
from collections import defaultdict
```

## Input Files as Tokens and Preprocessing

```python
with open('train.txt', 'r') as file:
    train = file.read()
train
```

```python
train = re.sub(r'\W',' ',train)
train
```

```python
train_token = train.lower().split()
train_token
```

```python
with open('val.txt', 'r') as file:
    val = file.read()
val
```

```python
val = re.sub(r'\W',' ',val)
val
```

```python
val_token = val.lower().split()
val_token
```

# Unsmoothed Unigram Probabilities

```python
un_uni_count = defaultdict(int)
un_uni_word_count = 0
for word in train_token :
    un_uni_count[word] = un_uni_count[word] + 1
    un_uni_word_count = un_uni_word_count + 1
```

```python
un_uni_word_count
```

```python
len(un_uni_count)
```

```python
un_uni_count
```

```python
un_uni_train_prob = defaultdict(int)
for word in un_uni_count :
    un_uni_train_prob[word] = un_uni_count[word]/un_uni_word_count
```

```python
un_uni_train_prob
```

# Unsmoothed Unigram Perplexity on the Training Dataset

```python
lap_uni_train_log_prob = defaultdict(int)
for word in un_uni_train_prob :
    lap_uni_train_log_prob[word] = math.log(un_uni_train_prob[word])
```

```python
lap_uni_train_log_prob
```

```python
log_prob = defaultdict(int)
sum_log_prob = 0
for word in un_uni_count :
    log_prob[word] = (-1)*(lap_uni_train_log_prob[word])*(un_uni_count[word])
    sum_log_prob = sum_log_prob + log_prob[word]
```

```python
perplexity = math.exp(sum_log_prob/un_uni_word_count)
perplexity
```
